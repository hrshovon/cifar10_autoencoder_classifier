# A short note on Autoencoder and Classifier Combination

<p>
An Autoencoder is an special type of neural network. in short, it tries to learn it's input by replicating it. It has two parts, one encoder part, which basically reduces the input data dimension and creates a compressed representation. The decoder part then expands it and produces the output. <BR/>
A classifier is well, what it's name implies, a machine learning algorithm that classifies stuff. In our case, its a neural network. Instead of using a stanalone convolutional neural network, we are taking the compressed representation from our autoencoder and feeding it to our classifier.
</p>

# Problem breakdown and strategy

This problem is an unbalanced dataset problem. So careful selection of metric is an essential thing here.<BR/>
Our dataset has 10 classes, with 5000 samples per class. if we discard 50% image from 3 of our classes, we still have 7 classes. So even if we get all of samples from those 3 classes wrong, we can still get up to 82% accuracy(70% if we had 5000 images per class for those 3 classes). So accuracy can only provide a very high level overview, not the complete picture.<BR/>
Therefore, we shall include <B>precision, recall and F1 score</B> to our metrics. after each training, we shall also print a heatmap to see confusion matrix in percentage form.

# Loss function consideration

<p>
We have two different type of output here. One output will be an image of same size, since we have an autoencoder. Another is a one hot encoded categorical output. For the categorical output, its fairly simple. We are simply using <I>"Categorical Crossentropy"</I>. <BR/><BR/>
For the autoencoder however, it's slightly tricky. <BR/><BR/>
before feeding the image into network, we normalize it. So 8 bit RGB representation turns into a 32bit floating point representation where each pixel has value of 0-1 for each channel. This is to ease up the convergence of neural network. <BR/><BR/>For Autoencoders, usually<B> input data = output data</B><BR/><BR/>
So our output will also be an image of same size and channels as input,having 3 channels per pixel and floating point value between 0 and 1 per channel. The output layer can be set to sigmoid to make sure it is always between 0 and 1. The loss function can be either: <BR/>
    <ul>
        <li>Binary crossentropy - to consider it as a pixelwise classification problem</li>
        <li>Mean Square Error/Mean Absolute Error - To consider it as a regression problem.</li>
    </ul>
Binary crossentropy is not going to be a very good loss function for this problem. Because:
     <ul>
        <li>It's not a simple binary pixel problem like mnist -where value is 0 or 1.</li>
        <li> Each of the pixels have their own intensity levels. So its a regression problem.</li>
    </ul>
So I chose <I>"mse(mean squared error)"</I>. 
</p>

# Some helper functions

```python
'''
Just a reverse search code for our class dictionary
'''
def key_from_value(value):
    for key,val in data_label_dict.items():
        if val==value:
            return key
    return -1

'''
first, let's write some helper functions/classes to ease up our work.
this function does the data stripping.
input 
    trainX -> training X data
    trainY -> training Y data
    classes_split_dict -> dictionary {class: split_fraction}
'''

def data_stripper(trainX,trainY,classes_split_dict):
    trainX_out=None
    trainY_out=None
    lst_idx=[]
    for key,v in classes_split_dict.items():
        class_label=key_from_value(key)
        split_frac=v
        del_item_idx=np.where(trainY==class_label)[0]
        class_label_random_samples=random.sample(list(del_item_idx),int(del_item_idx.shape[0]*split_frac))
        lst_idx.extend(list(class_label_random_samples))
    lst_idx=np.array(lst_idx)
    class_label_all=np.arange(0,trainY.shape[0])
    items_to_keep_idx=class_label_all[~np.isin(class_label_all,lst_idx)]
    trainX_out=trainX[items_to_keep_idx]
    trainY_out=trainY[items_to_keep_idx]
    return trainX_out,trainY_out    


'''
This is our custom generator function. It uses imblearn 
library's balanced data generator to generate balanced 
data from even the imbalanced dataset
'''

class BalancedDataGenerator(Sequence):
    """ImageDataGenerator + RandomOversampling"""
    def __init__(self, x, y, datagen, batch_size=32):
        self.datagen = datagen
        self.batch_size = batch_size
        self._shape = x.shape        
        datagen.fit(x)
        self.gen, self.steps_per_epoch = balanced_batch_generator(x.reshape(x.shape[0], -1), y, sampler=RandomOverSampler(), batch_size=self.batch_size, keep_sparse=True)

    def __len__(self):
        return self._shape[0] // self.batch_size

    def __getitem__(self, idx):
        x_batch, y_batch = self.gen.__next__()
        x_batch = x_batch.reshape(-1, *self._shape[1:])
        x,y=self.datagen.flow(x_batch, y_batch, batch_size=self.batch_size).next()
        return x,[x,y]


'''
This is our training class, it will take the train and test data, do KFold on the training data and provide result in the test 
this class is capable of doing the following:
if we provide it with a single model, it will assume that the model does both autoencoder+classification
if we provide it with two models, it will take the first model as autoencoder and the second as classifier.
we need to verify which one performs better.
'''


from datetime import datetime
class trainer_class:
    def __init__(self,sess_name,trainX,trainY,valX,valY,testX,testY,model1,model2=None,do_aug=False,steps_per_epoch=None, lossweights={"decoder": 1.0, "clf_out": 1.0}):
        self.sess_name=sess_name
        self.trainX=trainX
        self.trainY=trainY
        self.valX=valX
        self.valY=valY
        self.testX=testX
        self.testY=testY
        self.model1=model1
        self.model2=model2
        self.tempmodel=None
        self.fldr_name=self.sess_name+"/"
        self.do_aug=do_aug
        self.steps_per_epoch=steps_per_epoch
        self.loss_w=lossweights
    def setup_all(self):
        if os.path.exists(self.fldr_name)==False:
            os.mkdir(self.fldr_name)
        metrics = [
            "accuracy",
             ]
        if self.model2!=None:
            self.model1.compile(loss="mean_squared_error",optimizer=Adam(0.001))
            
        else:
            print("no model2 found. We shall use model1 for both autoencoder and classifier")
            losses={"decoder":"mse",
                          "clf_out":"categorical_crossentropy"}
            metrics_sm={"clf_out":metrics}
            self.model1.compile(optimizer=Adam(0.001), loss=losses, loss_weights=self.loss_w,metrics=metrics_sm)
    
    def train_it(self,epochs,bs):
        warnings.filterwarnings('ignore')
        train_X=self.trainX/255
        val_X=self.valX/255
        train_Y=to_categorical(self.trainY,num_classes=10)
        val_Y=to_categorical(self.valY,num_classes=10)
        histout=None
        if self.model2!=None:
            hist1=None
            hist2=None
            #first we train the AE as it's our feature extractor
            AE_filename=self.sess_name+"_AE.hdf5"
            CLF_filename=self.sess_name+"_CLF.hdf5"
            mc1=ModelCheckpoint(filepath=AE_filename,monitor="val_loss",save_best_only=True,verbose=1)
            rc1=ReduceLROnPlateau(patience=10,monitor="val_loss",factor=0.9, min_lr=0.00005, verbose=1,cooldown=1)
            es1=EarlyStopping(monitor="val_loss",patience=30)
            tb1=TensorBoard(log_dir=self.fldr_name+datetime.now().strftime("%Y%m%d-%H%M%S"))
            
            hist1=self.model1.fit(train_X,train_X,batch_size=bs,epochs=epochs,verbose=1,validation_data=(val_X,val_X),callbacks=[mc1,rc1,es1,tb1])
                
            self.model1=load_model(AE_filename)
            self.tempmodel=Model(self.model1.input,self.model1.get_layer("encoder").output)
            for layer,layer2 in zip(self.tempmodel.layers,self.model1.layers):
                layer.set_weights(layer2.get_weights())
                layer.trainable=False
            preds=self.tempmodel.predict(train_X)
            preds_val=self.tempmodel.predict(val_X)
            self.model2=clf(self.tempmodel,joined=False)
            metrics = [
            "accuracy",
             ]
            self.model2.compile(loss="categorical_crossentropy",optimizer=Adam(0.001),metrics=metrics)
            mc2=ModelCheckpoint(filepath=CLF_filename,monitor="val_loss",save_best_only=True,verbose=1)
            rc2=ReduceLROnPlateau(patience=10,monitor="val_loss",factor=0.9, min_lr=0.00005, verbose=1,cooldown=1)
            es2=EarlyStopping(monitor="val_loss",patience=30)
            tb2=TensorBoard(log_dir=self.fldr_name+ datetime.now().strftime("%Y%m%d-%H%M%S"))
            
            hist2=self.model2.fit(preds,train_Y,batch_size=bs,epochs=epochs,verbose=1,validation_data=(preds_val,val_Y),callbacks=[mc2,rc2,es2,tb2])
            histout = [hist1.history,hist2.history]
        else:
            #aug is only for this function though
            hist=None
            #we have a single model
            COMB_filename=self.sess_name+"_COMB.hdf5"
            
            rc=ReduceLROnPlateau(patience=10,monitor="val_loss",factor=0.9, min_lr=0.00005, verbose=1,cooldown=1)
            es=EarlyStopping(monitor="val_loss",patience=30)
            mc=ModelCheckpoint(filepath=COMB_filename,monitor="val_loss",save_best_only=True,verbose=1)
            tb=TensorBoard(log_dir=self.fldr_name+ datetime.now().strftime("%Y%m%d-%H%M%S"))
            hist=None
            if self.do_aug==False:

                hist=self.model1.fit(train_X,
                                       [train_X,train_Y],
                                       batch_size=bs,
                                       epochs=epochs,
                                       verbose=1,
                                       validation_data=(val_X,[val_X,val_Y]),
                                       callbacks=[rc,es,mc,tb])
            else:
                datagen = ImageDataGenerator(
                                                        featurewise_center=False,  # set input mean to 0 over the dataset
                                                        samplewise_center=False,  # set each sample mean to 0
                                                        featurewise_std_normalization=False,  # divide inputs by std of the dataset
                                                        samplewise_std_normalization=False,  # divide each input by its std
                                                        zca_whitening=False,  # apply ZCA whitening
                                                        zca_epsilon=1e-06,  # epsilon for ZCA whitening
                                                        rotation_range=0,  # randomly rotate images in the range (degrees, 0 to 180)
                                                        # randomly shift images horizontally (fraction of total width)
                                                        width_shift_range=0.1,
                                                        # randomly shift images vertically (fraction of total height)
                                                        height_shift_range=0.1,
                                                        shear_range=0.,  # set range for random shear
                                                        zoom_range=0.,  # set range for random zoom
                                                        channel_shift_range=0.,  # set range for random channel shifts
                                                        # set mode for filling points outside the input boundaries
                                                        fill_mode='nearest',
                                                        cval=0.,  # value used for fill_mode = "constant"
                                                        horizontal_flip=True,  # randomly flip images
                                                        vertical_flip=False,  # randomly flip images
                                                        # set rescaling factor (applied before any other transformation)
                                                        rescale=None,
                                                        # set function that will be applied on each input
                                                        preprocessing_function=None,
                                                        # image data format, either "channels_first" or "channels_last"
                                                        data_format=None)
                bgen = BalancedDataGenerator(train_X, train_Y, datagen, batch_size=bs)
                if self.steps_per_epoch==None:
                    steps_per_epoch = bgen.steps_per_epoch
                else:
                    steps_per_epoch = self.steps_per_epoch
                print(steps_per_epoch)
                hist=self.model1.fit_generator(
                                       bgen,
                                       steps_per_epoch,
                                       epochs=epochs,
                                       verbose=1,
                                       validation_data=(val_X,[val_X,val_Y]),
                                       callbacks=[rc,es,mc,tb])
            histout=[hist.history]
        with open(self.sess_name+"_training_logs.pkl","wb") as f:
            pickle.dump(histout,f)
        return histout
    
    '''
    This function outputs the confusion matrix of the test set and also prints the classification report
    '''
    
    def evaluate(self):
        warnings.filterwarnings('ignore')
        preds_clf=None
        testX=self.testX/255
        testY=to_categorical(self.testY,num_classes=10)    
        if self.model2!=None:
            AE_filename=self.sess_name+"_AE.hdf5"
            CLF_filename=self.sess_name+"_CLF.hdf5"
            self.model2=load_model(CLF_filename)
            #two models
            preds_ae=self.tempmodel.predict(testX)
            preds_clf=self.model2.predict(preds_ae)
        else:
            COMB_filename=self.sess_name+"_COMB.hdf5"
            self.model1=load_model(COMB_filename)
            preds_clf=self.model1.predict(testX)[1]
        preds_clf=[[np.argmax(preds_clf[i])] for i in range(preds_clf.shape[0])]
        conf_mat=confusion_matrix(self.testY,preds_clf)
        clf_rpt=classification_report(self.testY,preds_clf,target_names=data_labels)
        print(clf_rpt)
        return conf_mat

'''
This function takes in the history output of a model training and provides graphs 
'''
def print_comparative_graphs(dict_data,xlabel,ylabel,title,param_list):
    plt.figure(figsize=(15,8))
    for param in param_list:
        plt.plot(dict_data[param],label=param)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend(loc="upper_right")
    plt.title(title)
    plt.show()
'''
Given the name of the training session, it loads up the model and and returns confusion matrix
of the test set. It can also print or not print the classification report.
'''    
def evaluate_saved_model(task_name,testX,testY,single_model=True,print_clr=False):
        warnings.filterwarnings('ignore')
        preds_clf=None
        testX=testX/255
        if single_model==False:
            AE_filename=task_name+"_AE.hdf5"
            CLF_filename=task_name+"_CLF.hdf5"
            self.model2=load_model(CLF_filename)
            #two models
            preds_ae=self.tempmodel.predict(testX)
            preds_clf=self.model2.predict(preds_ae)
        else:
            COMB_filename=task_name+"_COMB.hdf5"
            model=load_model(COMB_filename)
            preds_clf=model.predict(testX)[1]
        preds_clf=[[np.argmax(preds_clf[i])] for i in range(preds_clf.shape[0])]
        conf_mat=confusion_matrix(testY,preds_clf)
        clf_rpt=classification_report(testY,preds_clf,target_names=data_labels)
        if print_clr==True:
            print(clf_rpt)
        return conf_mat
```

# Network

# Network 1


<p>
This network is slightly based on a paper(<A href="https://arxiv.org/abs/1412.6806">Striving for simplicity: The all convolutional Net</A>). I chose the parameters for the encoder layer from that paper, then made a Decoder Layer which resizes the and expands the encoded output to original image.
As for classifier, I played with different sizes and parameters. I was confused about the assignment requirements so I decided to keep a convolution layer for this network, and set the rest of the layers accordingly.
</p>

```python
def encoder(input_img):
    x=Conv2D(filters=96, kernel_size=(3,3))(input_img)
    x=Activation('relu')(x)
    x=Conv2D(filters=96, kernel_size=(3,3), strides=2)(x)
    x=Activation('relu')(x)
    x=Dropout(0.2)(x)
    x=Conv2D(filters=192, kernel_size=(3,3))(x)
    x=Activation('relu')(x)
    x=Conv2D(filters=40, kernel_size=(3,3), strides=2)(x)
    encoded=Activation('relu',name="encoder")(x)
    return encoded

def decoder(input_img,encoded):
    x = Conv2D(40,(3,3),padding="same")(encoded)
    x = Activation("relu")(x)
    x = UpSampling2D((2,2))(x)
    x = Conv2DTranspose(40, (3, 3))(x)
    x = Activation('relu')(x)
    x = Conv2DTranspose(192, (3, 3))(x)
    x = Activation('relu')(x)
    x = Conv2DTranspose(192, (3, 3))(x)
    x = Activation('relu')(x)
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(3, (3, 3), padding='same')(x)
    decoded = Activation('sigmoid',name="decoder")(x)
    model = Model(input_img, decoded,name="aec")
    return model

def autoencoder(shape):
    input_img = Input(shape=shape,name="input_img")
    encoded=encoder(input_img)
    autoencoded=decoder(input_img,encoded)
    return autoencoded

def clf(autoencoder,joined=False):
    input_shape=autoencoder.get_layer("encoder").output_shape[1:]
    ip=None
    if joined==False:
        ip=Input(shape=input_shape)
    else:
        ip=autoencoder.get_layer("encoder").output
        
    x=Conv2D(filters=32, kernel_size=(3,3), padding="same",name="clf_first_layer")(ip)
    x=Activation('relu',name="act1")(x)
    x=Dropout(0.5,name="drp1")(x)
    x=Flatten(name="flt_model")(x)
    x=BatchNormalization(name="bn1")(x)
    x=Dense(256,name="clf_dense1")(x)
    x=Activation('relu',name="act2")(x)
    x=Dense(10, activation="softmax",name="clf_out")(x)
    mdl_out=None
    if joined==True:
        mdl_out=Model(autoencoder.input,[autoencoder.output,x])
    else:
        mdl_out=Model(ip,x)
    return mdl_out
```

# New classifier, Adding regularizer

```python
def clf_2(autoencoder,joined=False):
    input_shape=autoencoder.get_layer("encoder").output_shape[1:]
    ip=None
    if joined==False:
        ip=Input(shape=input_shape)
    else:
        ip=autoencoder.get_layer("encoder").output
        
    x=Conv2D(filters=32, kernel_size=(3,3), padding="same",name="clf_first_layer",kernel_regularizer=l2(0.001))(ip)
    x=Dropout(0.25,name="drp1")(x)
    x=Activation('relu',name="act1")(x)
    x=Flatten(name="flt_model")(x)
    x=BatchNormalization(name="bn1")(x)
    x=Dense(256,name="clf_dense1",kernel_regularizer=l2(0.001))(x)
    x=Dropout(0.25,name="drp2")(x)
    x=BatchNormalization(name="bn2")(x)
    x=Activation('relu',name="act2")(x)
    x=Dense(100,name="clf_dense2",kernel_regularizer=l2(0.001))(x)
    x=Dropout(0.25,name="drp3")(x)
    x=BatchNormalization(name="bn3")(x)
    x=Activation('relu',name="act3")(x)
    x=Dense(10, activation="softmax",name="clf_out")(x)
    mdl_out=None
    if joined==True:
        mdl_out=Model(autoencoder.input,[autoencoder.output,x])
    else:
        mdl_out=Model(ip,x)
    return mdl_out
```

```python
def encoder(input_img):
    x=Conv2D(filters=96, kernel_size=(3,3))(input_img)
    x=Activation('relu')(x)
    x=Conv2D(filters=96, kernel_size=(3,3), strides=2)(x)
    x=Activation('relu')(x)
    x=Dropout(0.2)(x)
    x=Conv2D(filters=192, kernel_size=(3,3))(x)
    x=Activation('relu')(x)
    x=Conv2D(filters=40, kernel_size=(3,3), strides=2)(x)
    encoded=Activation('relu',name="encoder")(x)
    return encoded

def decoder(input_img,encoded):
    x = Conv2D(40,(3,3),padding="same")(encoded)
    x = Activation("relu")(x)
    x = UpSampling2D((2,2))(x)
    x = Conv2DTranspose(40, (3, 3))(x)
    x = Activation('relu')(x)
    x = Conv2DTranspose(192, (3, 3))(x)
    x = Activation('relu')(x)
    x = Conv2DTranspose(192, (3, 3))(x)
    x = Activation('relu')(x)
    x = UpSampling2D((2, 2))(x)
    x = Conv2D(3, (3, 3), padding='same')(x)
    decoded = Activation('sigmoid',name="decoder")(x)
    model = Model(input_img, decoded,name="aec")
    return model

def autoencoder(shape):
    input_img = Input(shape=shape,name="input_img")
    encoded=encoder(input_img)
    autoencoded=decoder(input_img,encoded)
    print(autoencoded.name)
    return autoencoded

def clf(autoencoder,joined=False):
    input_shape=autoencoder.get_layer("encoder").output_shape[1:]
    ip=autoencoder.get_layer("encoder").output
    x=Conv2D(filters=32, kernel_size=(3,3), padding="same")(ip)
    x=Activation('relu')(x)
    x=Dropout(0.5)(x)
    x=Flatten()(x)
    x=BatchNormalization()(x)
    x=Dense(256)(x)
    x=Activation('relu')(x)
    x=Dense(10, activation="softmax")(x)
    mdl_out=None
    if joined==True:
        mdl_out=Model(autoencoder.input,[autoencoder.output,x])
    else:
        mdl_out=Model(autoencoder.input,x)
    return mdl_out
aed=autoencoder((32,32,3))
mdl=Model(aed.input,aed.get_layer("encoder").output)
print(aed.summary())
clf=clf(mdl,joined=False)
print(clf.summary())
```

```python
def encoder_mod(input_img):
    x=Conv2D(filters=32, kernel_size=(3,3),padding="same",kernel_regularizer=l2(1e-4))(input_img)
    x=Activation('elu')(x)
    x=BatchNormalization()(x)
    x=Conv2D(filters=32, kernel_size=(3,3),padding="same",kernel_regularizer=l2(1e-4))(x)
    x=Activation('elu')(x)
    x=BatchNormalization()(x)
    x=MaxPooling2D(pool_size=(2,2))(x)
    x=Dropout(0.2)(x)
    
    x=Conv2D(filters=64, kernel_size=(3,3),padding="same",kernel_regularizer=l2(1e-4))(x)
    x=Activation('elu')(x)
    x=BatchNormalization()(x)
    x=Conv2D(filters=64, kernel_size=(3,3),padding="same",kernel_regularizer=l2(1e-4))(x)
    x=Activation('elu')(x)
    x=BatchNormalization()(x)
    x=MaxPooling2D(pool_size=(2,2))(x)
    x=Dropout(0.3)(x)
    
    x=Conv2D(filters=128, kernel_size=(3,3),padding="same",kernel_regularizer=l2(1e-4))(x)
    x=Activation('elu')(x)
    x=BatchNormalization()(x)
    x=Conv2D(filters=128, kernel_size=(3,3),padding="same",kernel_regularizer=l2(1e-4))(x)
    x=Activation('elu')(x)
    x=BatchNormalization()(x)
    x=MaxPooling2D(pool_size=(2,2))(x)
    x=Dropout(0.4)(x)
    
    encoded=Flatten(name="encoder")(x)
    return encoded

def decoder_mod(input_img,encoded):
    x=Reshape((4,4,128))(encoded)
    x=Conv2D(32,(3,3), kernel_regularizer=l2(1e-4),padding="same")(x)
    x=BatchNormalization()(x)
    x=Activation("elu")(x)
    x=UpSampling2D((2,2))(x)
    x=Conv2D(32,(3,3), kernel_regularizer=l2(1e-4),padding="same")(x)
    x=BatchNormalization()(x)
    x=Activation("elu")(x)
    x=UpSampling2D((2,2))(x)
    x=Conv2D(32,(3,3), kernel_regularizer=l2(1e-4),padding="same")(x)
    x=BatchNormalization()(x)
    x=Activation("elu")(x)
    x=UpSampling2D((2,2))(x)
    x=Conv2D(3,(3,3),padding="same")(x)
    decoded = Activation('sigmoid',name="decoder")(x)
    model = Model(input_img, decoded,name="aec")
    return model

def autoencoder_mod(shape):
    input_img = Input(shape=shape,name="input_img")
    encoded=encoder_mod(input_img)
    autoencoded=decoder_mod(input_img,encoded)
    return autoencoded
def clf_mod(autoencoder,joined=False):
    input_shape=autoencoder.get_layer("encoder").output_shape[1:]
    ip=None
    if joined==False:
        ip=Input(shape=input_shape)
    else:
        ip=autoencoder.get_layer("encoder").output
    x=Dense(10, activation="softmax",name="clf_out")(ip)
    mdl_out=None
    if joined==True:
        mdl_out=Model(autoencoder.input,[autoencoder.output,x])
    else:
        mdl_out=Model(ip,x)
    return mdl_out

def clf_mod_added_dense(autoencoder,joined=False):
    input_shape=autoencoder.get_layer("encoder").output_shape[1:]
    ip=None
    if joined==False:
        ip=Input(shape=input_shape)
    else:
        ip=autoencoder.get_layer("encoder").output
    x=Dense(512,kernel_regularizer=l2(1e-4))(ip)
    x=Activation("elu")(x)
    x=Dense(10, activation="softmax",name="clf_out")(x)
    mdl_out=None
    if joined==True:
        mdl_out=Model(autoencoder.input,[autoencoder.output,x])
    else:
        mdl_out=Model(ip,x)
    return mdl_out
```
